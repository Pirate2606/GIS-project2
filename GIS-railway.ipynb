```python
import geopandas as gpd
```

```python
import matplotlib.pyplot as plt
```

```python
dist = gpd.read_file(r"Shapefiles\districts.shp")
```

```python
dist.head()
```

```python
dist.plot(cmap="jet", edgecolor="black", column="district")
```

```python
AOI = gpd.read_file(r"Shapefiles\area_of_interest.shp")
```

```python
AOI.plot()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10,8))
dist.plot(ax=ax1, cmap="jet", edgecolor="black", column="district")
AOI.plot(ax=ax2, color="green")
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(10,8))
dist.plot(ax=ax1, cmap="jet", edgecolor="black", column="district")
AOI.plot(ax=ax2, color="green")
```

```python
atms = gpd.read_file(r"Shapefiles\atms.shp")
```

```python
fig, ax = plt.subplots(figsize=(10,8))
dist.plot(ax=ax, cmap="jet", edgecolor="black", column="district")
AOI.plot(ax=ax, color="none", edgecolor="black")
atms.plot(ax=ax, color="black")
```

```python
dist.crs
```

```python
dist = dist.to_crs(epsg=32629)
dist.plot(figsize=(10, 8))
```

```python
railway = gpd.read_file(r"Shapefiles\railway_tracks.shp")
```

```python
railway.plot()
```

```python
dist = dist.to_crs(epsg=4326)
fig, ax = plt.subplots(figsize=(10,8))
dist.plot(ax=ax, cmap="jet", edgecolor="black", column="district")
AOI.plot(ax=ax, color="none", edgecolor="black")
atms.plot(ax=ax, color="black")
railway.plot(ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(10,8))
intersection = gpd.overlay(dist, AOI, how="intersection")
intersection.plot(ax=ax, cmap="rainbow", edgecolor="black")
atms.plot(ax=ax, color="black")
```
